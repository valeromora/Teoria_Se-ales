```python
a = 1
while a:
  print("Choose an option:\n1. Book1 \n2. Book2 \n0. Exit\n\n")
  a = int(input())
  if a == 1:
    Book_data = pd.read_excel('Book1.xlsx')
  if a == 2:
    Book_data = pd.read_excel('Book2.xlsx')
  if a == 0:
    break

  print("Do you want to read data? \n1. Yes \n2. No \n0. Exit\n\n")
  a = int(input())
  if a == 1:
    print(Book_data)
  if a == 0:
    break

  print("Do you want to see synthesized information of wich column? \n1. Age \n2. Weight \n3. Height \n0. Exit\n\n")
  a = int(input())
  if a == 1:
    mean1 = Book_data['Edad'].mean()
    mode1 = Book_data['Edad'].mode()
    median1 = Book_data['Edad'].median()
    print("The mean is: ", mean1)
    print("The mode is: ", mode1)
    print("The median is:", median1)

  if a == 2:
    mean1 = Book_data['Peso'].mean()
    mode1 = Book_data['Peso'].mode()
    median1 = Book_data['Peso'].median()
    print("The mean is: ", mean1)
    print("The mode is: ", mode1)
    print("The median is:", median1)

  if a == 3:
    mean1 = Book_data['Estatura'].mean()
    mode1 = Book_data['Estatura'].mode()
    median1 = Book_data['Estatura'].median()
    print("The mean is: ", mean1)
    print("The mode is: ", mode1)
    print("The median is:", median1)
```
